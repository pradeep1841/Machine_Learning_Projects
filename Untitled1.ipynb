```python
#import some basic libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
dataset.head()
```

```python
## Divide the dataset into depenedent and indepenedent
X = dataset.iloc[:,3:13]
y = dataset.iloc[:,13]
```

```python
X.head()
```

```python
y.head()
```

```python
## Feature Engineering
geography = pd.get_dummies(X['Geography'],drop_first=True)
gender = pd.get_dummies(X['Gender'],drop_first=True)
```

```python
##Concatenate these variable with dataframe
X=X.drop(['Geography','Gender'],axis=1)
```

```python
X
```

```python
X=pd.concat([X,geography,gender],axis=1)
```

```python
#Splitting the dataset into training set ab Test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
X_train.shape
```

```python
## Part 2 Now lets create the ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, PReLU, ReLU
from tensorflow.keras.layers import Dropout
```

```python
## Lets initialize the ANN
classifier=Sequential()
```

```python
## Adding the input layer
classifier.add(Dense(units=11,activation='relu'))
```

```python
##Adding the first hidden layer
classifier.add(Dense(units=7,activation='relu'))
classifier.add(Dropout(0.2))
```

```python
## Second the second hidden layer
classifier.add(Dense(units=6, activation='relu'))
classifier.add(Dropout(0.3))
```

```python
## Adding the output layer
classifier.add(Dense(1, activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
## Ealry stopping
import tensorflow as tf
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    #start_from_epoch=0,
)
```

```python
model_history=classifier.fit(X_train, y_train, validation_split=0.33, batch_size=10, epochs=1000,callbacks=early_stopping)
```

```python
model_history.history.keys()
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel("accuracy")
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()

```

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel("loss")
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()

```

```python
#Predicting the Test set results
y_pred=classifier.predict(X_test)
y_pred=(y_pred>=0.5)
```

```python
# Make the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
# calculate the accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
